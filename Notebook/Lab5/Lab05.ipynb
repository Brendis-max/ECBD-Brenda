```python
#Aplicar estrategias de limpieza eliminación, transformación
#Eliminar duplicados
df = df.drop_duplicates()
```

```python
#S eliminaron registros duplicados para evitar datos repetids en el analisis, tambien se eliminaron filas con demasiados valores faltantes 
# en columnas mportantes como director y cast y los valores nulos en country y rating fueron remplazados por categorias generales.
```

```python
#Eliminar filas con demasiados valores nulos
df = df.dropna(subset=['director', 'cast'])
```

```python
#Rellenar valores faltantes
df['country'] = df['country'].fillna('Desconocido')
df['rating'] = df['rating'].fillna('No clasificado')
```

```python
df
```

```python
#Análisis descriptivo (CAPA DE INFORMACIÓN)
#1 Películas / Series
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="type")
plt.title("Cantidad de películas y series")
plt.xlabel("Tipo")
plt.ylabel("Cantidad")
plt.show()
```

```python
#CONCLUSION 1
#Netflix cuenta con una mayor cantidad de películas que de series.
```

```python
#2 Top 10 países con más contenido
top_paises = df['country'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_paises.index, y=top_paises.values)
plt.title("Top 10 países con más contenido")
plt.xlabel("País")
plt.ylabel("Cantidad")
plt.show()
```

```python
#CONCLUSION 2
#Estados Unidos aparece como el país con más contenido dentro del catálogo de Netflix.
```

```python
#3 Calificaciones del contenido
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="rating", order=df['rating'].value_counts().index)
plt.title("Clasificaciones de contenido")
plt.xlabel("Clasificación")
plt.ylabel("Cantidad")
plt.show()
```

```python
#4 Contenido lanzado en los últimos años
contenido_anio = df['release_year'].value_counts().sort_index().tail(20)
plt.figure(figsize=(12,5))
sns.barplot(x=contenido_anio.index, y=contenido_anio.values)
plt.title("Contenido lanzado en los últimos años")
plt.xlabel("Año")
plt.ylabel("Cantidad")
plt.show()
```

```python
##CONCLUSION 3
#La cantidad de contenido lanzado en Netflix aumentó considerablemente en los últimos años, mostrando el crecimiento de la plataforma mediante la incorporación constante de nuevas películas y series.
```
