```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 

from sklearn.metrics import (mean_absolute_error, mean_squared_error, root_mean_squared_error,r2_score)
```

```python
diabetes = load_diabetes(as_frame = True)
print (diabetes.DESCR)
```

```python
df = diabetes.frame
```

```python
df.head()
```

```python
df.describe()
```

## Seleccion de variablespredictoras y variables objetivas 
### Hipótesis 
Existe alguna relación entre las variables clinicas del paciente y la propagación de las diabetes 1 año despúes.

X = df.drop(columns="target")

```python
y = df["target"]
```

```python
X.head()
```

```python
y.head()
```

## Separación de datos para el entrenamiento y prueba

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state = 42
)
```

## Entrenamiento

```python
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

# Predicciones

```python
prediccion = modelo.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
comparacion = pd.DataFrame({
    "real": y_test.values,
    "prediccion": prediccion
}).reset_index(drop = True)
#Grafica
plt.figure(figsize = (12,6))
for i in range(len(comparacion)):
    plt.plot(
        [i,i],
        [comparacion["real"][i], comparacion["prediccion"][i]],
        color = "gray",
        alpha = 0.4
        
        
    )
plt.scatter(
        comparacion.index,
        comparacion["real"],
        label = "Valor real",
        marker = "o",
        s = 70
            
    )
plt.scatter(
        comparacion.index,
        comparacion["prediccion"],
        label = "Valor real",
        marker = "s",
        s = 70
            
    )
plt.xlabel("Paciente")
plt.ylabel ("Valor del target")
plt.title("Valores reales vs predichos")
plt.legend()
plt.grid(True)
plt.show()
```

## Metricas 

```python
# MAE
mae = mean_absolute_error(
    y_test, 
    prediccion
)
```

```python
print(mae)
```

```python
print(f"mae: {mae: .2f}")
```

```python
#MSE
mse = mean_squared_error(
    y_test,
    prediccion
)
```

```python
print(f"mse: {mse: .2f}")
```

```python
#RMSE
rmse = root_mean_squared_error(
    y_test,
    prediccion
)
```

```python
print(f"rmse: {rmse: .2f}")
```

```python
#Coheficiente de determinación (R2)
r2 = r2_score(
    y_test,
    prediccion
)
```

```python
print(f"r2: {r2: .2f}")
```
